```python
import pandas as pd
```

```python
ecom = pd.read_csv('https://raw.githubusercontent.com/MatinMasimli/DataScience_ML_Udemy_Exercises/master/Ecommerce_purchases/Ecommerce%20Purchases.csv')
```

**Check the head of the file**

```python
ecom.head()
```

**How many rows and columns are there**

```python
rows = len(ecom.index)
cols = len(ecom.columns)
print(rows, cols)
```

```python
ecom.info()
```

**What is the average purchase Price?**

```python
ecom['Purchase Price'].mean()
```

**What is the highest and lowest purchase price?**

```python
high = ecom['Purchase Price'].max()
low = ecom['Purchase Price'].min()
print("High: ", high)
print('Low: ', low)
```

**How many people have English 'en' as their language of choice on the website?**

```python
ecom[ecom['Language'] == 'en']['Language'].count()
```

**How may people have the job title of "Lawyer"?**

```python
# ecom[ecom['Job']=='Lawyer'].count()
# another way
len(ecom[ecom['Job']=='Lawyer'].index)
```

**How many people made the purchase during AM and how many people made the purchase during PM?**

```python
am = len(ecom[ecom['AM or PM'] == 'AM'])
pm = len(ecom[ecom['AM or PM'] == 'PM'])
print('AM: ', am)
print('PM: ', pm)
```

```python
""" another method 
 ecom['AM or PM'].unique() gives output
 array(['PM', 'AM'], dtype=object) 
 that means there r only 2 type of valuse am and pm
"""
ecom['AM or PM'].value_counts()
```

**What are the 5 most common Job Title?**

```python
vc = ecom["Job"].value_counts()
# value count sort in descending order
vc.head(5)
```

**Someone made a purchase that came from Lot: "90 WT", what was the Purchase Price for this transaction?**

```python
ecom[ecom['Lot'] == '90 WT']['Purchase Price']
```

**What is the email of the person with the following Credit Card Number: 4926535242672853**

```python
ecom[ecom['Credit Card'] == 4926535242672853]['Email']
```

**How many people have American Express as their Credit Card Provider and made a purchase above $90?**

```python
ecom[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)].count()
```

**How many people have credit card that expires in 2025?**

```python
ecom['CC Exp Date'].apply(lambda exp: exp[3:] == '25').sum()
```

**What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc..)**

```python
ecom['Email'].apply(lambda mail: mail.split('@')[1]).value_counts().head(5)
```
