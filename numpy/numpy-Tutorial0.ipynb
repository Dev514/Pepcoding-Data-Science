**Slicing**

Slicing is very similar to slicing in list.

Slicing list creates new list.

Slicing numpy array uses same array inside. Use **copy()** function to create new array insteed.

```python
ar = np.array([1,2,3,4,5,6,7])
b = ar[2:6]
b
```

```python
b[1] = 100
print(b)
print(ar)
```

```python
ar = np.array([1,2,3,4,5,6,7])
c = ar[2:6].copy()
c[1] = 200
print(c)
print(ar)
```

#**Split and Concatinate**

#**2D concatenate**

axis = 0, mean first axis(row)

axis = 1, mean second axis(col)

```python
grid = np.array([
                 [1,2,3,4],
                 [5,6,7,8]
])
z = np.concatenate([grid, grid])
z
```

```python
z = np.concatenate([grid, grid], axis=1)
z
```

**mixed dimensions**

vstack - vertically stack arrays(first axis)

hstack - horizontally stack arrays(second axis)

dstack - third axis

```python
x = np.array([1,2,3])
y = np.array([[1,2,3],
              [4,5,6]])
z = np.vstack([x,y])  # len(column) of all arrays must be same
z
```

```python
x = np.array([[1,2,3],
              [4,5,6]])
y = np.array([[10,10],
              [11,11]])
z = np.hstack([x,y])# len(row) of all arrays must be same
z
```

#**Operators**

numpy array operators with another array and with digits

```python
ar1 = np.random.permutation(np.arange(4))
ar2 = np.random.permutation(np.arange(4))
print(ar1)
print(ar2)
```

```python
# can do operation's with same shape

print(ar1 + ar2) # np.add(ar1, ar2)
print(ar1 - ar2)
print(ar1 * ar2)
print(ar1 / ar2)  # divide by 0 error
print(ar1 < ar2)  # True can be considered as 1 and False as 0
print(ar1 >= ar2)
```

##**broadcasting**

```python
print(ar1)
print(ar1 + 2)
print(ar1 - 2)
print(ar1 * 2)
print(ar1 / 2) # divide by 0 error
```

```python
# create new array with data coresponding to True
ar1[[True, False, False, True]] # can be used to filter data 
```
